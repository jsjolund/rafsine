```python
import matplotlib.pyplot as plt
import numpy as np

Tamb = 20
Told = Tamb
Tdelta1 = 80
Tdelta2 = 0

tau1 = 2
tau2 = 1
dt = 1
lmbda = 0.01

span = 100

assert(tau1+dt>tau2)
assert(tau1>tau2>0)

print(f'tau1/(tau1+dt) = {tau1/(tau1+dt)}')
print(f'dt/(tau1+dt)   = {dt/(tau1+dt)}')
print(f'tau2/(tau1+dt) = {tau2/(tau1+dt)}')
print(f'epsilon        = {tau2/tau1}')
```

```python
T = [Told]
for i in range(1, int(span/2)):
    Told = T[i-1]
    Tnew = tau1/(tau1+dt)*Told + dt/(tau1+dt)*(Tamb+(1-lmbda)*Tdelta1)
    T += [Tnew]
for i in range(int(span/2), span):
    Told = T[i-1]
    Tnew = tau1/(tau1+dt)*Told + dt/(tau1+dt)*(Tamb+(1-lmbda)*Tdelta2)
    T += [Tnew]
plt.plot(T)
plt.show()
```

```python
T = [Told]
for i in range(1, int(span/2)):
    Told = T[i-1]
    Tnew = Tamb + Tdelta1 + tau2/(tau1+dt)*(Told - Tamb - (1 - lmbda)*Tdelta1)
    T += [Tnew]
for i in range(int(span/2), span):
    Told = T[i-1]
    Tnew = Tamb + Tdelta2 + tau2/(tau1+dt)*(Told - Tamb - (1 - lmbda)*Tdelta2)
    T += [Tnew]
plt.plot(T)
plt.show()
```
