```python
def plot_rack(rack):
    rack['realpower_sum'] = (rack['realpower_l1'] + rack['realpower_l2'] + rack['realpower_l3']) / 1000
    xaxis = 'time'
    yaxis_real = ['realpower_sum',
            'front_temperature_temp_top',
            'front_temperature_temp_middle',
            'front_temperature_temp_bottom',
            'rear_temperature_temp_top',
            'rear_temperature_temp_middle',
            'rear_temperature_temp_bottom']
    yaxis_sim = ['realpower_sum',
            'sim_front_temperature_temp_top',
            'sim_front_temperature_temp_middle',
            'sim_front_temperature_temp_bottom',
            'sim_rear_temperature_temp_top',
            'sim_rear_temperature_temp_middle',
            'sim_rear_temperature_temp_bottom']
    fig, ax = plt.subplots(7, 1, figsize=(8, 16), sharex=True, sharey=False)
    plt.subplots_adjust(hspace=0)
    for a in ax:
        a.set_ylim((10, 49))
        a.xaxis.set_major_locator(mdates.HourLocator())
        a.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax[0].set_ylim((0, 20))
    ax[0].set_title(rack.name)
    ax[0].set_ylabel('Power (kW)')
    ax[1].set_ylabel('Front Top (C)')
    ax[2].set_ylabel('Front Mid (C)')
    ax[3].set_ylabel('Front Bottom (C)')
    ax[4].set_ylabel('Back Top (C)')
    ax[5].set_ylabel('Back Mid (C)')
    ax[6].set_ylabel('Back Bottom (C)')
    return rack, rack.name, xaxis, yaxis_real, yaxis_sim, ax

df, title, xaxis, yaxis_real, yaxis_sim, ax = plot_rack(rack05)
df.plot(x=xaxis, y=yaxis_real, ax=ax, rot=90, subplots=True, legend=False, x_compat=True, color='red')
df.plot(x=xaxis, y=yaxis_sim, ax=ax, rot=90, subplots=True, legend=False, x_compat=True, color='blue')
ax[0].legend(labels=['Exp.', 'LBM'])
```
