```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import pandas as pd

csv_path = '/home/user/code/rafsine/problems/pod2/input/'
```

```python
# Read csv files into Pandas DataFrames and replace NaN values
def open_csv(filename):
    content = pd.read_csv(csv_path + filename)
    content = content.interpolate(method='linear', limit_direction='both')
    return content

cooler01 = open_csv('cooler01.csv')
cooler02 = open_csv('cooler02.csv')
cooler03 = open_csv('cooler03.csv')
cooler04 = open_csv('cooler04.csv')
rack01 = open_csv('rack01.csv')
rack02 = open_csv('rack02.csv')
rack03 = open_csv('rack03.csv')
rack04 = open_csv('rack04.csv')
rack05 = open_csv('rack05.csv')
rack06 = open_csv('rack06.csv')
rack07 = open_csv('rack07.csv')
rack08 = open_csv('rack08.csv')
rack09 = open_csv('rack09.csv')
rack10 = open_csv('rack10.csv')
```

```python
def plot_rack(rack):
    rack['realpower_sum'] = rack['realpower_l1'] + rack['realpower_l2'] + rack['realpower_l3']
    xaxis = 'time'
    yaxis = ['realpower_sum', 
            'front_temperature_temp_top', 
            'front_temperature_temp_middle', 
            'front_temperature_temp_bottom', 
            'rear_temperature_temp_top', 
            'rear_temperature_temp_middle', 
            'rear_temperature_temp_bottom']
    fig, ax = plt.subplots(7, 1, figsize=(8, 16), sharex=True, sharey=False)
    plt.subplots_adjust(hspace=0)
    for a in ax[1:]:
        a.set_ylim((0, 49))
    ax[0].set_ylabel('Power (kW)')
    ax[1].set_ylabel('Front Top (°C)')
    ax[2].set_ylabel('Front Mid (°C)')
    ax[3].set_ylabel('Front Bottom (°C)')
    ax[4].set_ylabel('Back Top (°C)')
    ax[5].set_ylabel('Back Mid (°C)')
    ax[6].set_ylabel('Back Bottom (°C)')
    return rack, xaxis, yaxis, ax

df, xaxis, yaxis, ax = plot_rack(rack05)
df.plot(x=xaxis, y=yaxis, ax=ax, rot=90, subplots=True, legend=False, x_compat=True)

df, xaxis, yaxis, ax = plot_rack(rack06)
df.plot(x=xaxis, y=yaxis, ax=ax, rot=90, subplots=True, legend=False, x_compat=True)
```
